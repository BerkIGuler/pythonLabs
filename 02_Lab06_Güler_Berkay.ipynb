## Question 2
Download the files: catsM_1 to catsM_5.  These file contain data about cats and their body weight (in kg).
* catsM_1: contains the sex of 97 cats, 1: Female, 2:Male.
* catsM_2: contains the body weight(in kg)of the first 15 cats.
* catsM_3: contains the body weight(in kg) of the last 82 cats.



### Import numpy

```python
import numpy as np
```

### Load the data from the files above into numpy arrays.

```python
file1=np.loadtxt('catsM_1.csv',skiprows=1)
file2=np.loadtxt('catsM_2.csv',skiprows=1)
file3=np.loadtxt('catsM_3.csv',skiprows=1)
```

### Reshape all arrays, to have n rows, and 1 column. For example, the array containing the sex of all cats should have 97 rows and 1 column.  
#### The array containing the body weight of the first 15 cats should have 15 rows and 1 column.

```python
file1=file1.reshape((file1.size,1))
file2=file2.reshape((file2.size,1))
file3=file3.reshape((file3.size,1))
```

### Using vstack and hstack, create a table (97x2 array) that contains the columns: sex, body weight of all cats.

```python
body_weight=np.vstack((file2,file3))
table=np.hstack((file1,body_weight))
```

### Create a new array that contains the heart weights of all cats (in grams). Heart weight is approximately 25% of the body weight.

```python
hweight_in_g=body_weight*1/4*1000
```

### Reshape the heart weights to n rows, 1 column. 

```python
hweight_in_g=hweight_in_g.reshape(hweight_in_g.size,1)
```

### Join the heart weights as a new column to the array containing the sex and body weight of all cats.

```python
table=np.hstack((table,hweight_in_g))
```

### Using a single command display the heart weight  of all female cats 

```python
table[[table[:,0]==1]][:,2:3]
```

### Using vsplit and hsplit, create an array that contains the body weight and heart weights only of the first 15 cats.  Multiple statements may be used.
